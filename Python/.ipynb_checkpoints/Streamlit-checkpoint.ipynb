```python
import numpy as np
import pandas as pd
import plotly.express as px
import json
import webbrowser
import requests
import streamlit as st
from streamlit_option_menu import option_menu
from streamlit.components.v1 import components
```

```python
api_url = 'https://verified-sheep.dataos.app/mmclabtestdb/api/v1/mimic_labtest'
```

```python
response = requests.get(api_url).json()
df1 = pd.DataFrame(response)
df = df1.astype({"person_id": int
                 ,"test_date": object
                 ,"test_value_num":float
                 ,"total_test":int
                 ,"average_value":float
                 ,"percent_abnormal":float})
df['date'] = df['test_date'].astype('datetime64[ns]')
df['day'] = df.date.dt.date
```
